```python
#Transforming ca_df (median listing price dataset)
```

```python
#remove unnecessary columns
ca_df.drop(ca_df.columns[1:5], axis=1, inplace=True)
```

```python
ca_df.set_index("RegionName", inplace=True)
```

```python
#Reflect the DataFrame over its main diagonal by writing rows as columns and vice-versa
rev_df=ca_df.transpose()
```

```python
rev_df.head(10)
```

```python
#get the list of column names
col_names=list(rev_df.head(0)) 
print(col_names)
```

```python
#Replace missing values (NaN) with 0
rev_df.fillna(0, inplace=True)
```

```python
rev_df.head()
```

```python
#reset index column (for convenience) 
rev_df.reset_index(drop=False, inplace=True)
```

```python
rev_df.head()
```

```python
#calculate the number of rows
x=len(rev_df.index)
print(x)
```

```python
#calculate the number of columns
y=len(rev_df.columns)
print(y)
```

```python
#create an empty DataFrame
mp_df=pd.DataFrame(columns=["City","Date", "Value"])
```

```python
#slicing rev_df to fill out mp_df
for i in range(1,y):
    #slice column with dates + column with prices for each city
    ts=rev_df.iloc[:,[0,i]]
    city_l=[]
    b=col_names[i-1]
    #create a list with length=x. All elements of this list have the same value - city name
    for i in range(x):
        city_l.append(b)
    ts["City"]=city_l
    #rename columns in temporary DataFrame ts
    ts=ts.rename(columns={ts.columns[0]: "Date", ts.columns[1]: "Value", ts.columns[2]: "City"})
    #append temporary DataFrame ts to mp_df
    mp_df=mp_df.append(ts,ignore_index=True)
    ts=0
```

```python
mp_df.head()
```

```python
#start index from 1 instead of 0
mp_df.index += 1
#reset index
mp_df.reset_index(drop=False, inplace=True)
```

```python
mp_df.tail()
```

```python
#rename 'index' column to 'id'
mp_df.rename(columns={'index':'id'}, inplace=True)
```

```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
#Use declarative_base and create ORM classes for each table.
class Forecasts(Base):
    __tablename__='forecast'
    
    City=Column(String, primary_key=True)
    Forecast=Column(Float)
    
```

```python
class Price(Base):
    __tablename__='price'
   
    id=Column(Integer, primary_key=True)
    City=Column(String)
    Date=Column(String)
    Value=Column(Float)
```

```python
#Use the engine and connection string to create a database called hawaii.sqlite
engine = create_engine("sqlite:///housing.sqlite")
```

```python
#Once you have your ORM classes defined, create the tables in the database using create_all
Base.metadata.create_all(engine)
```

```python
#check tables names
engine.table_names()
```

```python
#connect to the database
conn=engine.connect()
```

```python
#call the to_sql method to convert df to a table in a database.
#set the 'keep_exists'='append' to  insert data into existing tabels.

forc_df.to_sql("forecast", conn, if_exists="append", index=False)
mp_df.to_sql("price", conn, if_exists="append", index=False)
```

```python
#We can then verify that everything works by querying the database
engine.execute("select * from forecast limit 5").fetchall()
```
