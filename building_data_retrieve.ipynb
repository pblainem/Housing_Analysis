```python

# Dependencies
import json
import requests
import pprint
import pandas as pd
```

```python

# connect to San Francisco Housing Data API
response = requests.get('https://data.opendatanetwork.com/resource/igjy-7nxp.json')
building_data = json.loads(response.text)

# print formatted JSON object
print(json.dumps(building_data, indent=2))


building_data   # How do I get more data points that 1000?
```

```python
# house_data_df = pd.read_json(response)

# city, buildingID, latitude, longitude = []
city = []
buildingID = []
buildingLat = []
buildingLon = []
buildingYear = []
squareFeet = []
buildingValue = []

# for building in building_data:
#     city.append(building[u"building_city"])
#     buildingLat.append(building[u"building_latitude"])
#     if ('building_built_year' in building):
#         buildingYear.append(building[u"building_built_year"])
#     if ('building_id' in building):    
#         buildingID.append(building[u"building_id"])
        
        
for building in building_data:
    if ('building_built_year' in building and 'building_id' in building and 'building_floor_area' in building and 'building_value' in building):
        buildingYear.append(building[u"building_built_year"])
        buildingID.append(building[u"building_id"])
        buildingLat.append(building[u"building_latitude"])
        city.append(building[u"building_city"])
        squareFeet.append(building[u"building_floor_area"])
        buildingLon.append(building[u"building_location"]["coordinates"][0])
        buildingValue.append(building[u"building_value"])
        



    
    
print(f"data contains {len(buildingYear)} year data points")
print(f"data contains {len(city)} city data points")
print(f"data contains {len(buildingLat)} latitude data points")
print(f"data contains {len(buildingLon)} longitude data points")
print(f"data contains {len(buildingID)} buildingID data points")
print(f"data contains {len(squareFeet)} square footage data points")
print(f"data contains {len(buildingValue)} building value data points")


```

```python
city_DF = pd.DataFrame(
{
    'City': city,
    'Year Built': buildingYear,
    'Latitude': buildingLat,
    'Longitude': buildingLon,
    'Square Footage': squareFeet,
    'Building Value': buildingValue

})

city_DF.head(900)
```
